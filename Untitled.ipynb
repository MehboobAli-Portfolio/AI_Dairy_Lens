```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, Subset
from torchvision import transforms
from PIL import Image, ImageFile
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Support for truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

# Data Preparation & Transform Wrapper

```python
class CattleDataset(Dataset):
    def __init__(self, root_dir):
        self.root_dir = root_dir
        self.images = []
        self.labels = []

        # 0: Healthy
        healthy_dirs = [
            os.path.join(root_dir, 'FMD_Cattle', '0'),
            os.path.join(root_dir, 'Lumpy Skin Images Dataset', 'Normal Skin'),
            os.path.join(root_dir, 'mastitis', 'normal teats')
        ]
        for d in healthy_dirs:
            if os.path.exists(d):
                for img_name in os.listdir(d):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        self.images.append(os.path.join(d, img_name))
                        self.labels.append(0)

        # 1: FMD
        fmd_dir = os.path.join(root_dir, 'FMD_Cattle', '1')
        if os.path.exists(fmd_dir):
            for img_name in os.listdir(fmd_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(fmd_dir, img_name))
                    self.labels.append(1)

        # 2: Lumpy Skin
        lumpy_dir = os.path.join(root_dir, 'Lumpy Skin Images Dataset', 'Lumpy Skin')
        if os.path.exists(lumpy_dir):
            for img_name in os.listdir(lumpy_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(lumpy_dir, img_name))
                    self.labels.append(2)

        # 3: Mastitis
        mastitis_dir = os.path.join(root_dir, 'mastitis', 'mastitis')
        if os.path.exists(mastitis_dir):
            for img_name in os.listdir(mastitis_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(mastitis_dir, img_name))
                    self.labels.append(3)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]
        return image, label

# NEW: Wrapper to apply specific transforms to our 80/20 splits
class TransformSubset(Dataset):
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform

    def __getitem__(self, index):
        image, label = self.subset[index]
        if self.transform:
            image = self.transform(image)
        return image, label

    def __len__(self):
        return len(self.subset)
```

# Model & Optimizer

```python
class CattleCNN(nn.Module):
    def __init__(self, num_classes=4):
        super(CattleCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 16 * 16, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

def get_loss_and_optimizer(model):
    loss_fn = nn.CrossEntropyLoss()
    # Added weight_decay for regularization (Normal Fit)
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    return loss_fn, optimizer
```

# Training & The Validation Graphs

```python
def train_and_validate(root_dir, model_save_path='model.pth', epochs=10):
    # 1. Transforms (Augmentation for Train, Clean for Val)
    train_transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    val_transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # 2. Split Dataset (80% Train, 20% Validation)
    full_dataset = CattleDataset(root_dir)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size

    # We use manual_seed(42) so the split is exactly the same every time you run it
    generator = torch.Generator().manual_seed(42)
    train_subset, val_subset = torch.utils.data.random_split(full_dataset, [train_size, val_size], generator=generator)

    train_dataset = TransformSubset(train_subset, transform=train_transform)
    val_dataset = TransformSubset(val_subset, transform=val_transform)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

    print(f"Total Images: {len(full_dataset)} | Training: {train_size} | Validating: {val_size}")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CattleCNN(num_classes=4).to(device)
    loss_fn, optimizer = get_loss_and_optimizer(model)

    # History tracking for graphs
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    print(f"Starting training on {device}...\n")

    for epoch in range(epochs):
        # --- TRAINING PHASE ---
        model.train()
        running_loss, correct_train = 0.0, 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = loss_fn(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct_train += torch.sum(preds == labels).item()

        train_loss = running_loss / len(train_loader)
        train_acc = (correct_train / train_size) * 100

        # --- VALIDATION PHASE (The 20% Test) ---
        model.eval()
        val_loss, correct_val = 0.0, 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = loss_fn(outputs, labels)

                val_loss += loss.item()
                _, preds = torch.max(outputs, 1)
                correct_val += torch.sum(preds == labels).item()

        val_loss = val_loss / len(val_loader)
        val_acc = (correct_val / val_size) * 100

        # Save metrics
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        print(f"Epoch {epoch+1:02d}/{epochs} | Train Loss: {train_loss:.4f}, Acc: {train_acc:.1f}% | Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.1f}%")

    # --- PLOT THE LEARNING CURVES ---
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss', color='blue')
    plt.plot(history['val_loss'], label='Validation Loss', color='red')
    plt.title('Loss Curve (Lower is Better)')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy', color='blue')
    plt.plot(history['val_acc'], label='Validation Accuracy', color='red')
    plt.title('Accuracy Curve (Higher is Better)')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    torch.save(model.state_dict(), model_save_path)
    print(f"\nModel saved successfully as {model_save_path}")
```

# Final Evaluation & Confusion Matrix

```python
def generate_final_report(root_dir, model_path='model.pth'):
    print("Loading model for Final Evaluation on the 20% Test Set...")

    # Load the exact same 20% validation split
    full_dataset = CattleDataset(root_dir)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    generator = torch.Generator().manual_seed(42)
    _, val_subset = torch.utils.data.random_split(full_dataset, [train_size, val_size], generator=generator)

    val_transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    val_dataset = TransformSubset(val_subset, transform=val_transform)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CattleCNN(num_classes=4)
    model.load_state_dict(torch.load(model_path, map_location=device, weights_only=True))
    model.to(device)
    model.eval()

    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    classes = ['Healthy', 'FMD', 'Lumpy Skin', 'Mastitis']

    # 1. Print Text Report
    print("\n" + "="*50)
    print("FINAL CLASSIFICATION REPORT")
    print("="*50)
    print(classification_report(all_labels, all_preds, target_names=classes))

    # 2. Plot Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix (20% Unseen Data)')
    plt.ylabel('True Disease (Reality)')
    plt.xlabel('Predicted Disease (AI Guess)')
    plt.show()
```

# The Deployment App Logic

```python
def predict_disease_smart(image_path, model_path='model.pth'):
    classes = ['Healthy', 'FMD', 'Lumpy Skin', 'Mastitis']

    if not os.path.exists(model_path):
        print(f"Error: Model not found at {model_path}")
        return

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CattleCNN(num_classes=4)
    model.load_state_dict(torch.load(model_path, map_location=device, weights_only=True))
    model.to(device)
    model.eval()

    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    try:
        image = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"Error loading image: {e}")
        return

    image_tensor = transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image_tensor)
        probabilities = torch.nn.functional.softmax(outputs[0], dim=0)

    print("\n" + "="*45)
    print(f"🩺 DIAGNOSIS REPORT: {os.path.basename(image_path)}")
    print("="*45)

    # Check all diseases instead of just picking the highest one
    detected_issues = []
    for i, cls in enumerate(classes):
        score = probabilities[i].item()
        # If the AI suspects a disease is more than 15% likely, add it to the report
        if score >= 0.15:
            detected_issues.append((cls, score))

    # Sort them so the highest percentage is at the top
    detected_issues.sort(key=lambda x: x[1], reverse=True)

    # Safety Check: Did nothing reach 15%? (Usually means it's a completely unrecognized image like a car or tree)
    if len(detected_issues) == 0:
        print("🚨 UNRECOGNIZED IMAGE: Please upload a clearer photo of the cow.")
    else:
        print("Conditions Detected:")
        for issue in detected_issues:
            if issue[0] == "Healthy" and issue[1] > 0.50:
                print(f"  ✅ {issue[0]} (Confidence: {issue[1]*100:.2f}%) - The cow appears fine.")
            else:
                print(f"  ⚠️ {issue[0]} (Confidence: {issue[1]*100:.2f}%)")

    print("\n(Raw Background Scores):")
    for i, cls in enumerate(classes):
        print(f"  - {cls}: {probabilities[i].item() * 100:.2f}%")
```
