```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from PIL import Image, ImageFile
import os
import sys

# Support for truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

# STEP 1: Data Preparation (Load & Process)

```python
class CattleDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.images = []
        self.labels = []
        
        # Mapping: 0: Healthy, 1: FMD, 2: Lumpy Skin, 3: Mastitis
        
        # 1. Healthy images
        healthy_dirs = [
            os.path.join(root_dir, 'FMD_Cattle', '0'),
            os.path.join(root_dir, 'Lumpy Skin Images Dataset', 'Normal Skin'),
            os.path.join(root_dir, 'mastitis', 'normal teats')
        ]
        for d in healthy_dirs:
            if os.path.exists(d):
                for img_name in os.listdir(d):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        self.images.append(os.path.join(d, img_name))
                        self.labels.append(0)

        # 2. FMD images
        fmd_dir = os.path.join(root_dir, 'FMD_Cattle', '1')
        if os.path.exists(fmd_dir):
            for img_name in os.listdir(fmd_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(fmd_dir, img_name))
                    self.labels.append(1)

        # 3. Lumpy Skin images
        lumpy_dir = os.path.join(root_dir, 'Lumpy Skin Images Dataset', 'Lumpy Skin')
        if os.path.exists(lumpy_dir):
            for img_name in os.listdir(lumpy_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(lumpy_dir, img_name))
                    self.labels.append(2)

        # 4. Mastitis images
        mastitis_dir = os.path.join(root_dir, 'mastitis', 'mastitis')
        if os.path.exists(mastitis_dir):
            for img_name in os.listdir(mastitis_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(mastitis_dir, img_name))
                    self.labels.append(3)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, label
```

# STEP 2 & 3: Model Architecture & Optimizer

```python
class CattleCNN(nn.Module):
    def __init__(self, num_classes=4):
        super(CattleCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 16 * 16, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

def get_loss_and_optimizer(model):
    loss_fn = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    return loss_fn, optimizer
```

# STEP 4 & 5: Training Loop & Save Function

```python
def train_model(root_dir, model_save_path='model.pth', epochs=10):
    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    dataset = CattleDataset(root_dir, transform=transform)
    dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
    
    # Get the total number of images for formatting
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    
    print(f"Total images loaded: {size}")
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CattleCNN(num_classes=4).to(device)
    loss_fn, optimizer = get_loss_and_optimizer(model)

    print(f"Starting training on {device}...\n")
    
    for epoch in range(epochs):
        print(f"Epoch {epoch+1}\n-------------------------------")
        model.train()
        
        running_loss = 0.0
        correct_predictions = 0
        
        for batch, (images, labels) in enumerate(dataloader):
            images, labels = images.to(device), labels.to(device)
            
            # Forward pass
            outputs = model(images)
            loss = loss_fn(outputs, labels)
            
            # Backpropagation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            # Record metrics
            running_loss += loss.item()
            _, predictions = torch.max(outputs, 1)
            correct_predictions += torch.sum(predictions == labels).item()
            
            # Print intermediate progress (e.g., every 10 batches)
            if batch % 10 == 0:
                loss_current = loss.item()
                # Calculate how many images we've processed so far
                current = batch * len(images) + len(images)
                # Ensure it doesn't print a number higher than the total size
                current = min(current, size) 
                print(f"loss: {loss_current:>7f}  [{current:>5d}/{size:>5d}]")
                
        # --- End of Epoch Summary ---
        avg_loss = running_loss / num_batches
        epoch_accuracy = (correct_predictions / size) * 100
        
        # Note: We label this "Training Error" because it is measuring the data it just studied.
        print(f"Training Error: \n Accuracy: {epoch_accuracy:.1f}%, Avg loss: {avg_loss:>8f} \n")

    torch.save(model.state_dict(), model_save_path)
    print(f"Step 5 Complete: Model saved as {model_save_path}")
```

# PREDICTION LOGIC (Deployment Phase)

```python

def predict_disease(image_path, model_path='model.pth'):
    classes = ['Healthy', 'FMD', 'Lumpy Skin', 'Mastitis']
    
    if not os.path.exists(model_path):
        print(f"Error: Model file '{model_path}' not found. Please run training first.")
        return

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CattleCNN(num_classes=4)
    model.load_state_dict(torch.load(model_path, map_location=device, weights_only=True))
    model.to(device)
    model.eval()

    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    image = Image.open(image_path).convert('RGB')
    image_tensor = transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image_tensor)
        probabilities = torch.nn.functional.softmax(outputs[0], dim=0)
    
    prediction = classes[torch.argmax(probabilities).item()]
    print(f"\n--- Result for: {os.path.basename(image_path)} ---")
    print(f"Prediction: {prediction}")
    print("Confidence Scores:")
    for i, cls in enumerate(classes):
        print(f"  - {cls}: {float(probabilities[i]*100):.2f}%")
```
